```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import warnings
warnings.filterwarnings('ignore')
```

```python
colors = ['#B7B7B7','#67BF7F', 'pink', '#F09C42', 'lightblue','#cde6c7','#68ACD5', '#FEDC5E', '#AAAAFF', '#ff7575']
colors1 = ['#B7B7B7','#67BF7F', 'pink', 'lightblue','#cde6c7','#68ACD5', '#FEDC5E', '#AAAAFF', '#ff7575']
datasets = ['Human_pancreatic_islets','Adam','10X_PBMC','mouse_ES_cell','human_kidney_counts','Macosko_mouse_retina','mouse_bladder_cell']
xs = ['Human_islet', 'Adam', '10X_PBMC','Mouse_ES', 'Human_kidney','Mouse_retina','Mouse_bladder']
```

```python
########   ARI    ##########
fig = plt.figure(figsize=(18, 7))
for dataset in datasets:
    dir0 = '../'
    ARI = []
    if dataset in ['Human_pancreatic_islets','Adam','10X_PBMC','mouse_bladder_cell']:
        methods = ['Seurat','PCA', 'ICA', 'ZIFA', 'VASC','scGAE','scGNN', 'DCA', 'scVI','scGraphNE']
        for method in methods:
            r = np.load(os.path.join(dir0,'results/clustering/{}/result_{}_{}.npz'.format(dataset,dataset,method)))
            ari = r['aril']
            ARI.append(ari)
        df = pd.DataFrame(ARI, index=methods)
        i = datasets.index(dataset)

        def plot_boxplot(a,b,ax):
            ax = plt.subplot(a,b,ax)
            bplot1 = ax.boxplot(df.values.T, labels=df.index,
                            vert=True,
                            medianprops={'color': 'k'},
                            flierprops={'markersize': 1},
                            patch_artist=True)

            for patch, color in zip(bplot1['boxes'], colors):
                patch.set_facecolor(color)

            y = MultipleLocator(0.2)
            ax.yaxis.set_major_locator(y)
            plt.ylim(0, 1)
            plt.xticks(fontsize=14, rotation=45, family='Arial', ha='right')
            plt.tick_params(labelsize=14)
            plt.grid(axis="y")
            plt.title("ARI of {}".format(xs[datasets.index(dataset)]),fontsize=20, fontproperties='Arial',loc="center")

    else:
        methods = ['Seurat', 'PCA', 'ICA', 'VASC', 'scGAE', 'scGNN', 'DCA', 'scVI', 'scGraphNE']
        for method in methods:
            r = np.load(os.path.join(dir0, 'results/clustering/{}/result_{}_{}.npz'.format(dataset, dataset, method)))
            ari = r['aril']
            ARI.append(ari)
        df = pd.DataFrame(ARI, index=methods)
        i = datasets.index(dataset)

        def plot_boxplot(a, b, ax):
            ax = plt.subplot(a, b, ax)
            bplot1 = ax.boxplot(df.values.T, labels=df.index,
                                vert=True,
                                medianprops={'color': 'k'},
                                flierprops={'markersize': 1},
                                patch_artist=True)

            for patch, color in zip(bplot1['boxes'], colors1):
                patch.set_facecolor(color)

            y = MultipleLocator(0.2)
            ax.yaxis.set_major_locator(y)
            plt.ylim(0, 1)
            plt.xticks(fontsize=14, rotation=45, family='Arial', ha='right')
            plt.tick_params(labelsize=14)
            plt.grid(axis="y")
            plt.title("ARI of {}".format(xs[datasets.index(dataset)]), fontsize=20, fontproperties='Arial',
                      loc="center")

    plot_boxplot(2,4,i+1)

plt.tight_layout(h_pad=2, w_pad=3)
plt.savefig('../figures/aribox_real.png',dpi=800, bbox_inches='tight')
plt.savefig('../figures/aribox_real.svg', dpi=300, format='svg', bbox_inches='tight')
plt.show()
```
