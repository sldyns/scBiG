```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import warnings
warnings.filterwarnings('ignore')
```

```python
datasets = ['10X_PBMC','Human_pancreatic_islets','Macosko_mouse_retina','mouse_ES_cell']
xs = ['Human PBMC', 'Human pancreas','Mouse retina','Mouse ES']
```

```python
########   ARI    ##########
fig = plt.figure(figsize=(12, 7))
for dataset in datasets:
    dir0 = '../'
    ARI = []
    if dataset in ['10X_PBMC','Human_pancreatic_islets']:
        methods = ['Seurat', 'ICA','ZIFA', 'graph-sc','scGAE','scGNN','DCA', 'scVI', 'SIMBA', 'scBiG']
        colors = ['#D9412B', '#F67948', '#FABB6E', '#F9E07F', '#4F9D9D', '#8f8ce7','#B7B7B7','#92C5DE', '#5C90C2', '#E79796']
        for method in methods:
            r = np.load(os.path.join(dir0,'results/clustering/{}/result_{}_{}.npz'.format(dataset,dataset,method)))
            ari = r['aril']
            ARI.append(ari)
        df = pd.DataFrame(ARI, index=methods)
        i = datasets.index(dataset)
        #print(df)
        aa=list(df.mean(axis=1).sort_values().index)
        rank=[]
        for j in methods:
            aa.index(j)
            rank.append(aa.index(j))
        rank, methods,colors = zip(*sorted(zip(rank, methods,colors)))
        methods=list(methods)
        colors=list(colors)
        df['Col']=df.mean(axis=1)
        df=df.sort_values(by ='Col',ascending=True)

        def plot_boxplot(a,b,ax):
            ax = plt.subplot(a,b,ax)
            bplot1 = ax.boxplot(df.values.T, labels=df.index,
                            vert=True,
                            medianprops={'color': 'k'},
                            flierprops={'markersize': 1},
                            patch_artist=True)

            for patch, color in zip(bplot1['boxes'], colors):
                patch.set_facecolor(color)

            y = MultipleLocator(0.2)
            ax.yaxis.set_major_locator(y)
            plt.ylim(0, 1)
            plt.xticks(fontsize=14, rotation=45, family='Arial', ha='right')
            plt.tick_params(labelsize=14)
            plt.grid(axis="y")
            plt.title("ARI of {}".format(xs[datasets.index(dataset)]),fontsize=20, fontproperties='Arial',loc="center")

    else:
        methods = ['Seurat', 'ICA','graph-sc','scGAE','scGNN','DCA', 'scVI', 'SIMBA', 'scBiG']
        colors = ['#D9412B', '#F67948', '#F9E07F', '#4F9D9D', '#8f8ce7','#B7B7B7','#92C5DE', '#5C90C2', '#E79796']
        for method in methods:
            r = np.load(os.path.join(dir0, 'results/clustering/{}/result_{}_{}.npz'.format(dataset, dataset, method)))
            ari = r['aril']
            ARI.append(ari)
        df = pd.DataFrame(ARI, index=methods)
        i = datasets.index(dataset)
        #print(df)
        aa=list(df.mean(axis=1).sort_values().index)
        rank=[]
        for j in methods:
            aa.index(j)
            rank.append(aa.index(j))
        rank, methods,colors = zip(*sorted(zip(rank, methods,colors)))
        methods=list(methods)
        colors=list(colors)
        df['Col']=df.mean(axis=1)
        df=df.sort_values(by ='Col',ascending=True)

        def plot_boxplot(a, b, ax):
            ax = plt.subplot(a, b, ax)
            bplot1 = ax.boxplot(df.values.T, labels=df.index,
                                vert=True,
                                medianprops={'color': 'k'},
                                flierprops={'markersize': 1},
                                patch_artist=True)

            for patch, color in zip(bplot1['boxes'], colors):
                patch.set_facecolor(color)

            y = MultipleLocator(0.2)
            ax.yaxis.set_major_locator(y)
            plt.ylim(0, 1)
            plt.xticks(fontsize=14, rotation=45, family='Arial', ha='right')
            plt.tick_params(labelsize=14)
            plt.grid(axis="y")
            plt.title("ARI of {}".format(xs[datasets.index(dataset)]), fontsize=20, fontproperties='Arial',
                      loc="center")

    plot_boxplot(2,2,i+1)

plt.tight_layout(h_pad=2, w_pad=3)
plt.savefig('../figures/aribox_real.svg', dpi=300, format='svg', bbox_inches='tight')
plt.show()
```
